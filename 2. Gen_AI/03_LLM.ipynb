# Large Language Model (LLM)

------------------------------------------------------------------------------------------------------

Large - Huge Dataset is being used

### Training LLM
* Stage 1 : Generative Pre-Training
* Stage 2 : Supervised Fine Tuning 
* Stage 3 : Reinforcement Learning Through Human Feedback

### Stage 1 : Generative Pre-Training

* The Huge internet data is trained on a transformer model which creates the Base GPT Model.
* Transformer model can generate only Text Summary, Sentiment Analysis, Sentence Completion, Translation. Since, Our aim is to create a chatbot, So we need to fine tune the Base GPT Model to create a conversational chat bot.

### Stage 2 : Supervised Fine Tuning (SFT)

* Step 1 : Create a carefully crafted conversations. These conversations are created by one human agent chatting with another human agent pretending to be a chatbot. The human pretending to be the chatbot will give the ideal response to each request.These Conversations will be used for training the data.

* Step 2 : The next step is to create the SFT training corpus, which involves using the conversation history as input and ideal response as output.

* Step 3 : Next, SFT training corpus is trained using Base GPT Model with stochastic gradient descent as its optimizer. Now, it will behave like a chat bot.

Note : LLM is not ready yet !.. This SFT chatbot model will be able to give the output based on the data it is trained on, but if asked about topics it hasn't been trained on, then the model will hallocinate and give you random answers..So we include Reinforcement learning through human feedback (Important step) , which increased the accuracy of model in tremendous way.

### Stage 3 : Reinforcement Learning Through Human Feedback

#### Step 1 :

* Now Human Agent chats with the SFT Trained chat bot model. The chatbot generates multiple answers to a question.



#### Step 2 :

Human Agent ranks those answers from best to worst (Top to bottom) based on :
* Accuracy (is it factually correct ? ),
* helpfullness (Does it address the user's request?) ,
* Safety ( Is the response appropriate and unbiased ),
* tone (Tone of response) .



#### Step 3 :

* These ranked responses are trained on the reward model.
* It learns to predict a reward score for any response based on past human preferences ( High Score = Great response!, Low Score = Needs improvement).
* Based on the learnings, it assigns a score to each response. A higher score means a better response.

#### Step 4 :

* We now take a copy of the SFT model and make a policy model. We improve this using PPO.
* policy model generates some answers(actions) for request(state).
* The reward model scores/rewards for each answers.
* Now the PPO model uses these scores to understand which responses are preferred by humans
* then PPO updates the policy model (decision making model) to produce high-scoring answers that alligns with human preferences(according to the reward model)
* However, PPO only allows small, controlled changes. If it makes too many drastic changes, it could go off track, so PPO carefully adjusts only a little bit at a time to improve.
* KL Divergence makes sure the model doesnt forget what it learned in earlier stages and prevents it from overfitting or drifting too far from good responses.
* ChatGPT improves by learning which answers people like, and it gets better at responding accordingly.
* The chatbot learns to avoid responses with low scores and aims for high scores.
