### 2. Encoder-Decoder vs Attention Mechanism vs Transformer

Encoder-Decoder :

* The words are sent one by one at each timestamp (Not Scalable)
*  The context vector is not sufficient to represent entire sentence If the sentence Increases, Bleau score dcreasing when length increases.
  
Attention Mechanism :

* The words are sent one by one at each timestamp (Not Scalable)
* Instead of single context, we provide additional context to our decoder, accuracy started increasing.

Transformer :

* It uses self attention module, which means all the words will be parallely sent to the encoder (Scalable)
* It uses contextual embedding.

### Encoder Architecture

####  STEPS: 

* The Input sentence is passed to the embedding layer and vectors are created for each word.
* These vectors are passed parallely to the self attention layer which creates the contextual vectors, which has the contextual meaning with other words in the sentence.
* These contextual vectors are passed to the feed forward neural network.
* The output vector of the neural network is passed on to the next encoder..

------------------------------------------------------------------------------------------------------

### Self Attention

* Self Attention, also known as scaled dot-product attention, is a crucial mechanism in the transformer architecture that allows the model to weigh the importance of different tokens in the input sequence relative to each other.

* The integral part of self attention is to convert the embedding into contextual embedding.

#### Query , Key, Value

* In normal embedding like Word2Vec, the same vectors is given for similar words eg : Apple phone and Apple fruit, both apple gets same embeddings, which destroys the contextual meaning.. so we convert the embedding into contextual embedding using self attention.

* Single head attention : single set pf k, q, v matrix

* multi head attention : muktiple set of k, q, v matrix

### why do we need multihead self attention

1. Helps to capture ambiguity inside the data. ( 1 sentence may have multiple perspective).. it helps to intepret one sentence in multiple way. eg: The chicken is ready to eat.
2. single attention was not able to understand the entire context of the data.
