### 1. What is Fine Tuning ?

Fine-tuning is the process of taking a pre-trained model and continuing training it on a specific, often smaller, dataset to adapt it to a new task or domain.



### 2. Types of Fine Tuning Techniques ?

* Full Parameter Fine Tuning
  
* LoRA - Low-Rank Adaptation of Large Language Models

* QLoRA - Quantized LoRA

### 3. Full Parameter Fine Tuning ?

* It updates all the model weights. eg : 175B weights needs to be updated.
  
* To update all the weights we will face resource constraints like GPU, RAM etc.

* In order to overcome this disadvantage we go for LoRA and QLoRA.

### 4. LoRA - Low-Rank Adaptation of Large Language Models

* pre-trained model, like a LLaMA or BERT. These models are massive with lots of parameters and expensive to fine-tune fully. This is where LoRA comes into the picture.

* So, LoRA stands for Low-Rank Adaptation of Large Language Models.

* The main idea of LoRA is: instead of fine-tuning the entire models, it freezes the original model weights. But track the changes of weights.

* If you are finetuning with LoRA then LoRA will track the new weights which will be as as size as pretrained weights then finally added with Lora tracked weights to give pretrained weights.

* LoRA doesn't update all the weight, but it uses a technique called matrix decomposition where it decomposes the matrix into 2 small trainable low rank matrices called  "LoRA Adaptors" into key layers (typically attention or feed-forward layers), and trains only those adaptors.

* Matrix decomposition is done based on Ranks. If we increase the rank the parameter keeps increasing

* This LoRA Technique reduces the number of trainable parameters drastically reduces memory and compute requirements.

  


### when to use high rank ? 

When the model needs to learn complex things like when it is not trained on something and you need to them to do a new thing..

### 5. QLoRA - Quantized LoRA

* In QLoRA, We quantize the original weights of the model from 16-bit or 32-bit to 4-bit numbers, which saves lot of memory.

* Then, like standard LoRA, we freeze the 4 bit model weights and inject trainable LoRA adapters (two small matrices A and B) into key layers.
