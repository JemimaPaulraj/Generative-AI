# LSTM - Long Short Term Memory and GRU

### LSTM

* LSTM is a another type of RNN
* It has memory cell. The main task of memory cell is to add what context is required and remove what context is not required.
* RNN has 1 feedback loop for short term memory and LSTM has 2 feedback loop, one for short term memory and one for long term memory

--------------------------------------------------------------------------------------------------------

### LSTM RNN Architecture

### Components

#### Memory Cell :
* It is for the long term memory.
* The main task of the long term memory cell is to add what context is required and remove what context is not required.

* C(t) = f(t) * C(t-1) + i(t) * C'(t)
  
* f(t) * C(t-1) --> Remove or forget some info, i(t) * C'(t) --> Add some info

Consider a Dataset !

### 1. Forget Gate

* Based on the Context the forget gate will let go some information and will not let go some information.

#### Step 1:
* At Time T = 2, The current input (xt) and previous hidden state (Ht-1) comes gets concatenated.

#### Step 2:
* The Concatenated vector is sent into the neural network with sigmoid activation function.

#### Step 3:
* The resultant output f(t) is point wise multiplied with previous memory cell state (ct-1).

--------------------------------------------------------------------------------------------------------------------------------------------------------

### 2. Input Gate and Candidate memory

* Based on the context the Input Gate checks if any information has to be added to the memory cell.

### Step 1 :

* I(t) and c(t) is calculated in the same way as the f(t).
* I(t) uses sigmoid activation function and the c(t) uses Tanh activation function.

#### Step 2:
* I(t) and C(t) is point wise multiplied to check if any new information needed to be added in the memory cell.

### Step 3:

* Adding new information to the memory cell using poni wise point wise addition operation.

--------------------------------------------------------------------------------------------------------------------------------------------------------

### 3. Output Gate

### Step 1 :

* O(t) is calculated is the same way like f(t)

### Step 2:

* The Tanh function is applied on the vectors of C(t) and point wise multiplied with O(t).
* The resultant vector H(t) will be the short term memory and current memory cell after forgetting and adding some information will be the long term memory c(t).

Note : Back propogation is done to reduce the loss by adusting the weights Wi, wc, wo

### Summary

* Forget Gate determines how much past information should be discarded.
* Input Gate decides what new information should be stored in the cell state.
* Cell State Update modifies the memory by combining retained and new information.
* Output Gate regulates what part of the memory should be sent as the hidden state.

### Disadvantages of LSTM

* The architecture is complex.

* Lots of parameter inside the architecture.

* Training time is very high for huge data. so they created GRU


-------------------------------------------------------------------------------------------------------------------------------------------------------

# GRU

* GRU is the simplified version of LSTM RNN.
* Less parameters.
* Training time is less.

### GRU Architecture

### Which One to Use?
* Use LSTM when handling long sequences or tasks requiring memory over many time steps (e.g., speech recognition, language modeling).
* Use GRU when computational efficiency is important or when working with small datasets or intermediate dataset.
* Use RNN for small dataset.

### Note :

* LSTM is Best for large dataset
* RNN is best for small dataset
* If it is a intermediate dataset try between LSTM and GRU

-------------------------------------------------------------------------------------------------------------------------------------------------------

### Few Difference Between LSTM and GRU

1. Number of Gates : LSTM - 3, GRU - 2
2. Memory : LSTM - Cell state, Hidden state, GRU - Hidden State
3. Parameter Count : LSTM - More , GRU - Less
4. Computational complexity : LSTM - More , GRU - Less
5. Performance : LSTM - Best, GRU - Okay
