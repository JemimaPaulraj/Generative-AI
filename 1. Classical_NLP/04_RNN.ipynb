# RNN (Recurrent Neural Network)

------------------------------------------------------------------------------------------------------

* We use RNN for sequential data(data that has an order, like time series, text, or speech).
* RNNs have memory, meaning they remember previous inputs while processing the next ones.
* In case of sequential data we need to pass one word at a time to our neural network.

  eg : The Food is good, at t=1 send "The" , at t=2 send "Food", at t=3 send "is", at t=4 send "good".
* The hidden neurons will contain the context of the previous word. At the end of the sentence the neural network will contain the context of all the previous words.

------------------------------------------------------------------------------------------------------

Lets consider a dataset !

### Step 1: Process the First Word ("The")

T=1

* The first word's one-hot encoded vector x11 (10000) enters the hidden layer.
* The input is multiplied with the initial weight W, added to the bias b, and passed through the activation function, giving output O1.
* O1 is stored in the hidden neuron for context retention.

  ##### O1 = f ( x11 * w + b )

### Step 2:  Process the Second Word ("food") 

T=2

* The next words one-hot encoded vector x12 (01000) enters the hidden layer.
* Again the input is multiplied with the same initial weight and added with bias.
* Now the RNN uses both the current processed input x12 and the previous hidden state output O1 to calculate a new O2.
* ' helps decide how much of the past is carried forward to the next step.
* O1 is stored in the hidden neuron for context retention.

    ##### O2 = f (x12 * w + O1 * W' + b)

### Step 3:  Process the Third Word ("Taste") 

T = 3

* The third words one-hot encoded vector x13 (00100) enters the hidden layer.
* The RNN uses x13 and the previous hidden state output O2 to calculate a new O3.

    ##### O3 = f (x13 * w + O2 * W' + b)

### Step 4:  Process the Fourth Word ("Taste") 

T = 4

* The fourth words one-hot encoded vector x14 (00010) enters the hidden layer.
* The RNN computes O4 using x14 and O3.

    ##### O4 = f (x14 * w + O3 * W' + b)

### Step 5: Final Output

* After processing the whole sentence, The final output is decided by the activation function, then the loss is calculated.
* If the loss is high, backward propogation is done to minimize the loss.

--------------------------------------------------------------------------------------------------------------------------------------------------------

## Steps In RNN:

* The vectors of the each words will be passed one by one at each time stamp into the hidden layer.

* The input vector is multiplied with the initial weight and added with bais and given to the activation function, which gives O1

                     O1 = f ( x11 * w + b )

* Then this O1 will gets transferred back to the same hidden layer (feedback loop) to retain the context of previous words, that assisgns the weight w`

* ' helps decide how much of the past is carried forward to the next step. Without W' the RNN will not know what to remember and forget.

* Now the next word's vector comes in, now the RNN uses both the current processed input x12 and the previous hidden state output O1 to calculate a new O2, and then the same step continues for rest of the word.

                   O2 = f (x12 * w + O1 * W' + b)

* After processing the entire sentence, the final output is determined by the activation function. Then, the loss is calculated, and if it is high, backpropagation is performed to adjust the weights and minimize the loss.

### Problems in Simple RNN:

* It has short Term Memory

* Vanishing and exploding gradient problem

Vanishing Gradient problem:

* eg - Sigmoid: When using the sigmoid activation function, its derivative is always between 0 and 1. During backpropagation, as gradients propagate backward through multiple layers, they keep getting multiplied by small values, eventually becoming very close to zero. This leads to the vanishing gradient problem, where weight updates become negligible, preventing the model from effectively learning and converging to the global minimum.

Exploding gradient problem:

* eg - ReLU: The ReLU activation function allows gradients to pass through for positive values, but for negative values, the gradient becomes zero (due to ReLU being zero for negative inputs). This can cause neurons to become inactive, leading to the dying ReLU problem. Additionally, large gradient updates can sometimes cause instability, making it harder for the model to converge smoothly to the global minimum.

# Types of RNN

* One to One RNN
* One to Many RNN
* Many to one RNN
* Many to Many RNN

### 1. One to One RNN

* 1 input and 1 output
* eg: Sequence prediction (predicting next word based on the previous word)

### 2. One to Many RNN

* 1 input and many output
* eg : Image Captioning (cat is sitting on table)

### 3. Many to One RNN

* Many input and 1 output
* eg: Image search (cat eating food), sentiment analysis

### 4. Many to Many RNN

* Many input and Many output
* eg: Language Translation

# Bidirectional RNN

eg : Jemima Eats ------- In Chennai

In order to answer this, it also needs the future context.. This can be solved by bidirectional RNN.

* A Bidirectional RNN processes input sequences in both forward and backward directions using two separate hidden layers, improving context understanding. 
* This allows the model to capture dependencies from both past and future inputs, making it useful for tasks like fill in the blanks etc...
