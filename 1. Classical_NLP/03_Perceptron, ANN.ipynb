##### Vanishing gradient problem

* The value of sigmoid ranges from 0 to 1, above 0.5 is considered as 1 and below 0.5 is considered as 0.
* The derivative of sigmoid activation function always ranges from 0 to 0.25

For the below example if we use sigmoid activation function:

* During back propogation, we can see sigmoid derivative is used in every step.

* The value of derivative will always between 0 to 0.25, so the value keeps on decreasing throughout the end. so we will be getting a very small value after the multiplication.

What happens if we get a small value ?

* When updating the weight, since the value is very small, At one point of time the new weight will be approximately equal to old weight. It will hardly change. So weights are not getting updated.


* The situation when the weights are not getting updated or it is just getting updated by a smaller value, this problem is called as vanishing gradient problem.
* This can be solved by using other activation function
